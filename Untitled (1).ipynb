```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
df['Sleep_Disorder_encoded'] = le.fit_transform(df['Sleep Disorder'].fillna('No_Disorder'))


features = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']
X = df[features]
y = df['Sleep_Disorder_encoded']


rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)


importances = rf.feature_importances_
for f, imp in zip(features, importances):
    print(f"{f}: {imp:.4f}")


import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.barh(features, importances)
plt.title("Feature Importances for Sleep Disorder Prediction")
plt.show()

```
